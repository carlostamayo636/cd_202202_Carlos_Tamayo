```python
import matplotlib.pyplot as plt
from kneed import KneeLocator
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df3)
scaled_features[:5]
```

```python
kmeans = KMeans(
  init="random",
  n_clusters=12,
  n_init=10,
  max_iter=300,
  random_state=42
)
```

```python
kmeans.fit(scaled_features)
```

```python
# Cuadrado de errores estándar más bajo (SSE)
kmeans.inertia_
```

```python
# Posición final de los centroides
kmeans.cluster_centers_
```

```python
# Número de iteraci´nes requeridas
kmeans.n_iter_
```

```python
kmeans.labels_
```

```python
kmeansLabels = pd.DataFrame(kmeans.labels_)
```

Metodo del codo

```python
kmeans_kwargs = {
  "init": "random",
  "n_init": 10,
  "max_iter": 300,
  "random_state": 42,
}

sse = []
for k in range(1, 21):
  kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
  kmeans.fit(scaled_features)
  sse.append(kmeans.inertia_)
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(1, 21), sse)
plt.xticks(range(1, 21))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

```python
kl = KneeLocator(
  range(1, 21), sse, curve="convex", direction="decreasing"
)
kl.elbow
```
